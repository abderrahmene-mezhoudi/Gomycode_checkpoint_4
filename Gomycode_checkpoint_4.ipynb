## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
x=HdataDf["sqft_living"].values[:,np.newaxis] 
y=HdataDf["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

### Applying Linear regression to our training set

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
model=LinearRegression()   
model.fit(x_train,y_train)
```

### Ploting the linear regression

```python
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

### Measuring the performance of linear regression using the testing set

```python
predicted=model.predict(x_test)
```

```python
Diff_1 = y_test - predicted
g=plt.plot((Diff_1), marker='.', linestyle='-' )
```

Here we have do a simple error calculate ,by subtracting predicted values from the real ones and we can constat that in general the results are null ( except some outliers ). From this plot we can say that our model is working as good as expected.

```python
D1 = metrics.r2_score(y_test,predicted)
print("R squared 1 =", D1)
```

We can say that half of the ouput variation can be explained by the model's inputes
